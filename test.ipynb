```python
def check_user_status(user_uid="0141432593019719"):
    # 1) Load the same Bearer token you saved earlier
    token = pathlib.Path("token").read_text().strip()

    # 2) Compose the URL
    url = f"https://prod-api02.swasthyaingit.in/prod/srs/api/v1/Connection/checkUserStatus/{user_uid}"

    # 3) Copy the key headers that the browser sends
    headers = {
        "Authorization": f"Bearer {token}",
        "Accept": "application/json, text/plain, */*",
        "Origin": "https://v2.swasthyaingit.in",
        "Referer": "https://v2.swasthyaingit.in/",
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/138.0.0.0 Safari/537.36"
        ),
        # Optional CORS / fetch hints  usually not required
        "sec-fetch-site": "same-site",
        "sec-fetch-mode": "cors",
        "sec-fetch-dest": "empty",
        'sec-ch-ua': '"Not)A;Brand";v="8", "Chromium";v="138", "Google Chrome";v="138"',
        'sec-ch-ua-platform': '"Windows"',
        'sec-ch-ua-mobile': '?0',
    }

    # 4) Make the request
    r = requests.get(url, headers=headers, timeout=10)
    r.raise_for_status()            # raise if 4xx/5xx

    return r.text
```
